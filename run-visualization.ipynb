```python
base = pd.read_csv('splits/mpc_full_feats_scaled_4927_4.csv')
base.drop(columns=['comp', 'bandgap'], inplace=True)
base
```

```python
df['kind'] = 'adv'
base['epoch'] = -10
base['batch'] = -10
base['kind'] = 'base'

combo = pd.concat([df, base]).reset_index(drop=True)

for col in ('kind',):
    combo[col] = pd.Categorical(combo[col])

combo.dropna(inplace=True)
combo
```

```python
from sklearn.decomposition import PCA
from umap import UMAP

pca = PCA(n_components=2)
umap = UMAP(n_neighbors=70, n_components=2, metric='manhattan', init='pca')

numeric = combo.iloc[:, :147]
combo[['pca_x', 'pca_y']] = pca.fit_transform(numeric.values)
combo[['umap_x', 'umap_y']] = umap.fit_transform(numeric.values)
```

```python
sns.relplot(combo, x='pca_x', y='pca_y', hue='epoch', ec=None, palette='rho_sequential', s=5, legend='brief', alpha=0.2)
```

```python
sns.relplot(combo, x='umap_x', y='umap_y', hue='epoch', ec=None, palette='rho_sequential', s=5, legend='brief', alpha=0.4)
```

```python
xcomp = pd.Series(pca.components_[0], index=numeric.columns)
xcomp.sort_values()
```

```python
scales = pd.read_csv('mpc_scales.csv')
mean = scales['mean'].values
scale = scales['scale'].values
```

```python
if combo.iloc[:, :147].mean().max() < 10:
    combo.iloc[:, :147] *= scale.T
    combo.iloc[:, :147] += mean.T
```

```python
from scipy.stats import epps_singleton_2samp

kstests = pd.Series(index=numeric.columns)

samp1 = combo.query('kind == "adv"')
samp2 = combo.query('kind == "base"')

noise1 = np.random.randn(*samp1.index.shape) * 1e-1
noise2 = np.random.randn(*samp2.index.shape) * 1e-1
for col in numeric.columns:
    kstests.loc[col] = epps_singleton_2samp(samp1[col] + noise1, samp2[col] + noise2).statistic

kstests.sort_values().tail(20)
```

```python
fig, axs = plt.subplots(4, 4, figsize=(20, 20), sharex=True, sharey=False)
axs = axs.flatten()
i = 0

seen_data = set()
for col in kstests.sort_values().tail(30).index[::-1]:
    data = ' '.join(col.split(' ', maxsplit=1)[1:]) if ' ' in col else col
    if data not in seen_data:
        seen_data.add(data)
        ax = axs[i]
        ax.hexbin(data=combo, x='epoch', y=col, cmap='rho_heatmap', gridsize=20, mincnt=1)
        ax.set_title(col)
        i += 1
        if i == len(axs):
            break
```
