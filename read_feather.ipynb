```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
nom_bandgap = pd.read_feather("SAL_nom_bandgap_training_loss.feather")
nom_deltae  = pd.read_feather("SAL_nom_delta_e_training_loss.feather")

raw_bandgap = pd.read_feather("SAL_raw_bandgap_training_loss.feather")
raw_deltae  = pd.read_feather("SAL_raw_delta_e_training_loss.feather")
```

```python
loss_columns = [col for col in nom_bandgap.columns if 'epoch' not in col]
```

```python
xsh=['Xshift_tsne','Xshift_umap']
```

```python
dfs = [nom_bandgap, nom_deltae]#, raw_bandgap, raw_deltae]


# Plot 'Xshift_tsne' and 'Xshift_umap' from each DataFrame on the same plot
for i, df in enumerate(dfs):
    plt.semilogy(df['Xshift_tsne'], label=f'{df.columns[0]} - Xshift_tsne')
    plt.semilogy(df['Xshift_umap'], label=f'{df.columns[1]} - Xshift_umap')

plt.xlabel('Time')
plt.ylabel('Xshift Values')
plt.title('Line Plot of Xshift Values from Multiple DataFrames')
plt.legend()
plt.show()
```

```python

# Set a larger figure size
plt.figure(figsize=(19, 10))
for loss_column in loss_columns:
    plt.semilogy(df['epoch'], df[loss_column], label=loss_column)
#plt.plot(df['epoch'], df)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss Over Time')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
# Set a larger figure size
plt.figure(figsize=(19, 10))
plt.boxplot([df[col] for col in loss_columns], labels=loss_columns, vert=False, sym="")  # Using vert=False for horizontal box plots

plt.xscale('log')  # Logarithmic scale on the x-axis
plt.xlabel('Logarithmic Loss')
plt.title('Box Plot of Multiple Losses')
plt.show()

```
